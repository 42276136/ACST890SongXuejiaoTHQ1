```python
## part (a) import pandas module and save as pd
import pandas as pd
## change the file name to student_results.csv
## read this file from Downloads folder
dataset = pd.read_csv('~/Downloads/student_results.csv')
## display the table that shows all the data in student_results.csv
dataset
```

```python
## part (b) fill all nan value with zero
dataset = dataset.fillna(0)
```

```python
## part (c) calculate each student's total mark
import matplotlib.pyplot as plt
import numpy as np
## let A be the weight of each assessment 
A = np.asarray([[0.07],
               [0.03],
               [0.03],
               [0.02],
               [0.03],
               [0.02],
               [0.024],
               [0.016],
               [0.07],
               [0.03],
               [0.04],
               [0.02],
               [0.6]])
B = dataset.iloc[:,1:14]
## adding Total.marks to the final columns of the table
dataset['Total.marks']=np.dot(B, A)
dataset.head()

```

```python
## part (d) boxplot for domestic and international students based on Total.marks
dataset.boxplot(column='Total.marks', by='Student category')
```

```python
## part (d) the mean of total marks in each class is shown in Total.marks which is the last column.
dataset.groupby("Class No")["Total.marks"].mean()  
## part (d) the standard deviation of total marks in each class is shown in Total.marks which is the last column.
dataset.groupby("Class No")["Total.marks"].std()
## Total.marks shows the standard deviation of each group 
dataset.groupby("Class No")["Total.marks"].var()

##put in a table:
n = len(dataset)
Stat_table = dataset.loc[0:n, ['Total.marks','Class No']].groupby('Class No').agg({'mean', 'std', 'var'})
print('Stat_table:\n', Stat_table)
```

```python
##part(e) assign grade based on Total.marks
Grade = []
for s in dataset["Total.marks"]:
    if s < 50:Grade.append('F')
    elif s < 65:Grade.append('P')
    elif s < 75:Grade.append('Cr')
    elif s < 85:Grade.append('D')
    else:Grade.append('HD')
dataset['Grade']=Grade
dataset.head()
```
